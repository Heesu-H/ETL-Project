# Check called API data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Load csv made from api call

```python
file_url = '01_Outputs/melb_suburbs_types.csv'
df = pd.read_csv(file_url)
df
```

# Plot suburbs a place counts

```python
fig, ax = plt.subplots(figsize=(40,20))
labels = df["suburb"]
xvalue1 = df["No liquor_store"]
xvalue2 = df["No police"]
xvalue3 = df["No park"]
ax.bar(labels, xvalue3)
ax.bar(labels, xvalue1)
ax.bar(labels, xvalue2)
plt.xticks(rotation=90)

fig.savefig('/path/to/figure.pdf')

plt.show()
```

Graph shows API results used as counts caps out at 20.
